```python
%matplotlib inline

import matplotlib.pyplot as plt
import pandas as pd

plt.style.use('ggplot')
```

```python
df=pd.read_csv('train_loan.csv')
```

```python
df.head()
```

```python
df.keys()
```

```python
df=df.dropna()
```

```python
code={'N':0,'Y':1}
for k,i in df.iterrows():
    if i['Loan_Status']==i['Loan_Status']:
        df.loc[k,"Loan_Status"]=code[i['Loan_Status']]
```

```python
code={'Female':0,'Male':1}
for k,i in df.iterrows():
    df.loc[k,"Gender"]=code[i['Gender']]
```

```python
code={'No':0,'Yes':1}
for k,i in df.iterrows():
    df.loc[k,"Self_Employed"]=code[i['Self_Employed']]
```

```python
code={'Not Graduate':0,'Graduate':1}
for k,i in df.iterrows():
    df.loc[k,"Education"]=code[i['Education']]
```

```python
code={'No':0,'Yes':1}
for k,i in df.iterrows():
    df.loc[k,"Married"]=code[i['Married']]
```

```python
code={'Urban':2,'Semiurban':1,'Rural':0}
for k,i in df.iterrows():
    df.loc[k,"Property_Area"]=code[i['Property_Area']]
```

```python
df.loc[df.Dependents == '3+', ['Dependents']] = 3
```

```python
import numpy as np
A=df2.as_matrix([df2.columns[:]])
X=df2.as_matrix([df2.columns[[6,7]]])
```

```python
# X[:,1]=np.log(X[:,1]+100)
```

```python

from sklearn.cluster import DBSCAN
db = DBSCAN(eps=1.5, min_samples=20).fit(X)
```

```python
db.labels_.max()
```

```python
ax=plt.figure(figsize=(15,10))
# plt.rcParams['figure.figsize'] = (12,9)
plt.xlabel("CoapplicantIncome")
plt.ylabel("LoanAmount(k)")
plt.title("Loan")
# plt.yscale('log')
color_list=['r','g','b','c','m','y','k']
for i in xrange( -1,db.labels_.max()+1):
    is_loan=(A[:,11]==1) & (db.labels_==i)
    plt.scatter(X[is_loan,0], X[is_loan,1], marker='o',color=color_list[i])
    


    
    is_not_loan=(A[:,11]==0) & (db.labels_==i)
    plt.scatter(X[is_not_loan,0], X[is_not_loan,1], marker='x',color=color_list[i])
```
